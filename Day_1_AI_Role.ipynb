```python
import numpy as np


data = np.array([[25, 40000, 85],
                 [45, 85000, 92],
                 [30, 52000, 78],
                 [50, 90000, 96]], dtype=float)

min = data.min(axis = 0)
max = data.max(axis = 0)

normalized = (data - min) / (max - min)
print(normalized)
```

***Why is NumPy faster than Python lists?***

Numpy is faster because it uses contigous array to store data which is easier and faster to setch and use, whereas Python uses pointer based array allocation where each data of array is squeezed into some part of memory and is connected through pointers.

***What is broadcasting?***

Broadcasting is when we give uneven arrays to perform operation and numpy takes care of it using a memory efficient strategy rather than blind copying of data again and again. There are 2 rules that are used for broadcasting.

* Arrays must be equal
* One of the dimension should be 1 and not any other



***When can broadcasting become dangerous?***

Dangers in broadcasting includes inefficient usage of memory when handling large high dimensional data, which causes memory to fill up easily. Making this inefficient, and a efficient way to process this kind of data would be to chunk the calculation parts to manageable parts that do not block memory.


***Explain reshaping without changing data.***

Reshaping produces a copy of the orignal array in the requested shape but does not change the original one, changing values in reshaped one does not affect the original.


***Difference between flatten() and ravel().***

* Flatten ==> Converts the given n-D array to 1-D array by making a copy of it in some memory face.

* Ravel ==> Converts a given n-D array to 1-D array by making a view of the original array
